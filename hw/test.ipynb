```python
import numpy as np
import pandas as pd
import random
import matplotlib.pyplot as plt
```

```python
delta_t = 1/500
T = 2
N = int(T/delta_t)
np.random.seed(1)
Bt = np.cumsum(np.random.normal(0.0, delta_t**0.5, size=N))
plt.plot(np.arange(N)*delta_t, Bt)
plt.title("Brownian Motion Simulation")
plt.xlabel("x")
plt.ylabel("y")
```

```python
np.random.seed(1)
def BMsimulation(delta_t=1/500, T=2, sim_n=1000):
    N = int(T/delta_t)
    sim_result = np.zeros((sim_n, N))
    for i in range(sim_n):
        Bt = np.cumsum(np.random.normal(0.0, delta_t**0.5, size=N))
        sim_result[i, :] = Bt
    return sim_result

delta_t=1/500
T=2
sim_n=1000
sim_result = BMsimulation(delta_t, T, sim_n)
```
